```python
def extract_url(url_x):
    parsed_uri=urlparse(url_x)
    result='{uri.scheme}://{uri.netloc}/'.format(uri=parsed_uri)
    return (result)
```

```python
dataset['website']=dataset['URL'].apply(lambda x: extract_url(x))
```
